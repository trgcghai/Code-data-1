```python
from math import sqrt

# Example usage of sqrt and squaring
example_value = 4
sqrt_value = sqrt(example_value)
squared_value = example_value ** 2


import math
def GaussianProb(x_value, g_mean, g_std): #Nhớ check ảnh chụp đt.
    x1 = 1/math.sqrt(2*math.pi*(g_std**2))
    exp1 = math.exp(-0.5*((x_value - g_mean)/g_std)**2)
    return x1 * exp1

```

```python
#Check ảnh chụp
#pclassX (số lận xuất hiện x trên tâp)
#plcassY (số lần xuất hiện y trên tập)
#pclassZ (số lần xuất hiện z trên tập)

pclassX = (2/7)*(GaussianProb(6.9, 4.75, 0.212132))*(GaussianProb(3.1, 3.0500, 0.070711))
pclassX = pclassX/(GaussianProb(6.9, 5.928571, 0.875051))*(GaussianProb(3.1, 3.085714, 0.167616))
print(pclassX)

pclassY = (2/7)*(GaussianProb(6.9, 6.7, 0.424264))*(GaussianProb(3.1, 3.05000, 0.212132))
pclassY = pclassY/(GaussianProb(6.9, 5.928571, 0.875051))*(GaussianProb(3.1, 3.085714, 0.167616))

print(pclassY)
pclassZ = (3/7)*(GaussianProb(6.9, 6.2, 0.3))*(GaussianProb(3.1, 3.133333, 0.230940))
pclassZ = pclassZ/(GaussianProb(6.9, 5.928571, 0.875051))*(GaussianProb(3.1, 3.085714, 0.167616))
print(pclassZ)

import numpy as np
maxvalue = np.max([pclassX, pclassY, pclassZ])
print('Max  Prob = ', maxvalue)
print(np.sum([pclassX, pclassY, pclassZ]))
```

```python
X_train = df_train[['ti', 'ci']].values #Explanatory variable (feature)
y_train = df_train[['drug']].values #Target variable    (resonpse)

X_train = X_train.astype(float)
```

```python
import pandas as pd

df = pd.read_csv('Iris.csv')
df
```

```python
df_train, df_test = train_test_split(df, test_size=0.8, random_state=16)
```

```python
X_train = df_train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values #Explanatory variable (feature)
y_train = df_train[['Species']].values #Target variable    (resonpse)

X_train = X_train.astype(float)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
X_test = df_test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values #Explanatory variable (feature)
y_test = df_test[['Species']].values #Target variable    (resonpse)

X_test = X_test.astype(float)
```

```python
y_pred = gnb.predict(X_test)
y_pred
```

```python
#compare
from sklearn import metrics

# Reshape y_train and y_test to be 1-dimensional arrays


print("Gaussian Naive Bayes model accuracy(in %):", metrics.accuracy_score(y_test, y_pred)*100)
```
