```python
import pandas as pd
column_name = ['Id', 'Name','Age','Weight','m0006','m0612','m1218','f0006','f0612', 'f1218']
df = pd.read_csv('patient_heart_rate.csv', names=column_name, header=0)
df
```

```python
df[['FirstName', 'LastName']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)

df
```

```python
weight = df['Weight']
for i in range(0, len(weight)):
    x = str(weight[i])
    if 'lbs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x / 2.2)
        weight[i] = y
    if 'kgs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        weight[i] = x

df.rename(columns={'Weight': 'Weight_kgs'}, inplace=True)
df
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
df = df.drop_duplicates(subset=['FirstName', 'LastName', 'Age', 'Weight_kgs'])
df
```

```python
df.FirstName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
df.LastName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(df)
```

```python
df['Age'].fillna(df['Age'].mode()[0], inplace=True)
df
```

```python
df['Weight_kgs'].fillna(df['Weight_kgs'].median(), inplace=True)
df
```

```python
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight_kgs', 'FirstName', 'LastName'], value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'FirstName', 'LastName'])

df_temp = df['gender_time'].str.extract("(\D)(\d+)(\d{2})", expand=True)

df_temp.columns = ['Gender', 'Lower_hour', 'Upper_hour']
df = pd.concat([df, df_temp], axis=1)
df = df.drop(['gender_time'], axis=1)
df
```

```python
import numpy as np 
df = df.replace('-', np.nan).dropna(subset=['PulseRate'])
df
```

```python
df['FirstName'].isnull().sum()
df['LastName'].isnull().sum()
df['FirstName'].fillna('Unknown', inplace=True)
df['LastName'].fillna('Unknown', inplace=True)
df
```

```python
df = df.reset_index()
df
```

```python
df.to_csv('patient_heart_rate_clean.csv')
```

```python
# Bieu do pie the hien ti le benh nhan nam va nu
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('patient_heart_rate_clean.csv', header=0, delimiter=',', encoding='utf-8')

df.head(10)

gb = df.groupby(['Gender'])['Gender'].agg(['count'])

labels = gb.index
data = list(gb['count']) 

colors = sns.color_palette('pastel') 

plt.pie(data, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True)

plt.show()
```

```python
# tao cot FullName = FirstName + ' ' + LastName
df['FullName'] = df['FirstName'] + ' ' + df['LastName']
df.head(10)
```

```python
# bieu do so sanh do tuoi giua cac ten benh nhan
plt.figure(figsize=(20, 20))
sns.barplot(x='FullName', y='Age', data=df)
plt.show()
```

```python
# bieu do so sanh nhip tim trung binh cua cac benh nhan
plt.figure(figsize=(20, 20))
df.groupby(['FullName'])['PulseRate'].agg(['mean']).reset_index()
sns.barplot(x='FullName', y='PulseRate', data=df)
plt.show()
```

```python
# bieu do so sanh nhip tim theo tung nhom gio co phan loai gioi tinh
plt.figure(figsize=(11, 11))
df['TimePeriod'] = df['Lower_hour'].astype(str) + '-' + df['Upper_hour'].astype(str)
df.groupby(['TimePeriod'])['PulseRate'].agg(['mean']).reset_index()
sns.barplot(x='TimePeriod', y='PulseRate', hue='Gender', data=df)
plt.show()
```

```python
# bieu do phan tan the hien su tuong quan giua trung binh do tuoi va trung binh nhip tim
sns.lmplot(data=df, x='Age', y='PulseRate', fit_reg=True)
plt.show()
```

```python
# bieu do duong the hien su thay doi nhip tim trung binh cua cac benh nhan theo khung gio 00-06, 06-12, 12-18 theo tung nhom gioi tinh
sns.lineplot(x='TimePeriod', y='PulseRate', hue='Gender', data=df)
plt.show()
```
