```python
# Sắp xếp DataFrame theo cột 'M1' theo thứ tự tăng dần.
df.sort_values(by='M1', ascending=True)
```

```python
# Sắp xếp DataFrame theo cột 'M1' theo thứ tự tăng dần và cột 'M2' theo thứ tự giảm dần.
df.sort_values(by=['M1','M2'], ascending=[True, False])
```

```python
# Lọc các hàng trong DataFrame mà giá trị của cột 'M1' lớn hơn hoặc bằng 5.
df[df['M1'] >= 5]
```

```python
# Lọc các hàng mà giá trị của cột 'M1' lớn hơn 6 và cột 'M2' lớn hơn 5.
df[(df['M1'] > 6) & (df['M2'] > 5)]
```

```python
# Lọc các hàng trong DataFrame mà giá trị của cột 'KT' là 'C'.
df.loc[df['KT'] == 'C']
```

```python
# Tính toán tổng hợp các giá trị tối thiểu và tối đa của cột 'M1', và giá trị trung bình và tổng của cột 'M2'.
df.aggregate({'M1': ['min', 'max'], 'M2': ['mean', 'sum']})
```

```python
# Đếm số lượng giá trị theo từng khu vực ('KV').
df.groupby(['KV'])['KV'].agg(['count'])
```

```python
# Tính toán số lượng, giá trị nhỏ nhất, trung bình, độ lệch chuẩn, lớn nhất và tổng của cột 'M1' theo từng khu vực ('KV').
df.groupby(['KV'])['M1'].agg(['count', 'min', 'mean', 'std', 'max', 'sum'])
```

```python
# Tính toán giá trị nhỏ nhất và trung bình của cột 'M1' và 'M2' theo từng khu vực ('KV').
df.groupby(['KV'])[['M1','M2']].agg(['min', 'mean'])
```

```python
# Tính toán giá trị nhỏ nhất và lớn nhất của cột 'M1' và 'M2' theo khu vực ('KV') và khu vực thi ('KT').
df.groupby(['KV','KT'])[['M1','M2']].agg(['min', 'max'])
```

```python
# Tạo một DataFrame mới bằng cách nhóm theo khu vực ('KV') và đếm số lượng bản ghi.
df_gr = df.groupby(['KV'], as_index=False).count()
df_gr
```

```python
# Tạo một bảng xoay (pivot table) với các giá trị của 'M1' và 'M2', nhóm theo cột 'KV', và tính toán giá trị nhỏ nhất, trung bình, lớn nhất.
pd.pivot_table(df, values=['M1','M2'], columns='KV', aggfunc=['min','mean','max'])
```

```python
# Tạo một bảng xoay (pivot table) mới với các giá trị của 'M1' và 'M2', nhóm theo cột 'KV' và 'KT', và tính toán giá trị nhỏ nhất, trung bình, lớn nhất.
pd.pivot_table(newdf, values=['M1','M2'], columns=['KV','KT'], aggfunc=['min','mean','max'])
```
