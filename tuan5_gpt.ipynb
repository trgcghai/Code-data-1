```python
import pandas as pd

# Tạo một DataFrame với hai cột 'Math' và 'Eng'
df = pd.DataFrame({'Math': [4, 4, 7, 8, 12, 13, 15, 18, 22, 23, 23, 25],
                   'Eng': [2, 5, 4, 7, 7, 8, 5, 4, 5, 11, 13, 8]})
# Hiển thị kích thước của DataFrame
df.shape
```

```python

# Chia cột 'Math' thành 5 nhóm theo tỷ lệ phần trăm
df['bin1'] = pd.qcut(df['Math'], q=5)
df

# Đếm số lượng các giá trị trong mỗi nhóm của 'bin1'
df['bin1'].value_counts()

# Chia cột 'Math' thành 5 nhóm với các tỷ lệ phần trăm cụ thể
df['bin2'] = pd.qcut(df['Math'], q=[0, .2, .4, .6, .8, 1])
df

# Chia cột 'Math' thành 5 nhóm với các tỷ lệ phần trăm cụ thể và gán nhãn cho mỗi nhóm
df['bin3'] = pd.qcut(df['Math'],
                           q=[0, .2, .4, .6, .8, 1],
                           labels=['A', 'B', 'C', 'D', 'E'])
df
```

```python
# Chia cột 'Eng' thành các nhóm với các mức 'Bad', 'OK', 'Good'
df['bin4'] = pd.cut(df['Eng'],
                      bins=[0, 5, 8, float('Inf')],
                      labels=['Bad', 'OK', 'Good'])
df

# Đếm số lượng các giá trị trong mỗi nhóm của 'bin4'
df['bin4'].value_counts()
```

```python
import pandas as pd

# Tạo một DataFrame với hai cột 'GDP' và 'Countries'
gdp = [11, 22, 33, 44, 55, 66, 77]
countries = ['Spain', 'France', 'Spain', 'Germany', 'France']
df = pd.DataFrame(list(zip(gdp, countries)),
                  columns=['GDP', 'Countries'])
df
```

```python
from sklearn  import preprocessing

# Mã hóa các giá trị chuỗi trong cột 'Countries' thành các giá trị số
label_encoder = preprocessing.LabelEncoder()
df['labelCountries']= label_encoder.fit_transform(df['Countries']) 
df
```

```python
# Sử dụng One Hot Encoding để biến đổi cột 'Countries' thành nhiều cột nhị phân
dfOneHotEncoding = pd.get_dummies(df.Countries, prefix='Country')
print(dfOneHotEncoding.head())

# Ghép DataFrame ban đầu với DataFrame sau khi sử dụng One Hot Encoding
df = pd.concat([df, dfOneHotEncoding], axis=1)
df
```

```python
# Sử dụng LabelBinarizer để biến đổi cột 'Countries' thành các cột nhị phân
from sklearn.preprocessing import LabelBinarizer
y = LabelBinarizer().fit_transform(df.Countries)
y
```

```python
# Sử dụng OneHotEncoder để biến đổi cột 'Countries' thành các cột nhị phân
from sklearn .preprocessing import OneHotEncoder
y = OneHotEncoder().fit_transform(df[['Countries']]).toarray()
print(y)
```

```python
# Tạo một DataFrame với ba cột 'Year', 'HCM' và 'LA'
year = [2010,2011,2012,2013,2014,2015]
salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]
salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]
df = pd.DataFrame(list(zip(year, salaryHCM, salaryLA)),
                  columns=['Year', 'HCM', 'LA'])
df
```

```python
# Tìm giá trị tuyệt đối lớn nhất của cột 'LA'
max_la = df['LA'].abs().max()
# Chuẩn hóa cột 'LA' bằng cách chia cho giá trị lớn nhất
df['macLA'] = df['LA'] / max_la
df
```

```python
# Sử dụng MaxAbsScaler để chuẩn hóa cột 'HCM' và 'LA'
from sklearn .preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM', 'LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM', 'LA']].columns)
print(scaled_df)
```

```python
# Sử dụng MaxAbsScaler để biến đổi một mẫu mới
vSc = scaler.transform([[4.5, 5.75]])
vSc
```

```python
# Sử dụng MaxAbsScaler để đảo ngược biến đổi một mẫu đã được chuẩn hóa
vR = scaler.inverse_transform([[0.5, 0.8]])
vR
```

```python
# Tính giá trị min và max của cột 'LA'
min_la = df['LA'].min()
max_la = df['LA'].max()
# Chuẩn hóa cột 'LA' sử dụng Min-Max Scaling
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)
df
```

```python
# Sử dụng MinMaxScaler để chuẩn hóa cột 'HCM' và 'LA'
from sklearn .preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM', 'LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM', 'LA']].columns)
print(scaled_df)
```

```python
# Tính giá trị trung bình và độ lệch chuẩn của cột 'LA'
mean_la = df['LA'].mean()
std_la = df['LA'].std()
# Chuẩn hóa cột 'LA' sử dụng Z-Score
df['zLA'] = (df['LA'] - mean_la) / std_la
df
```

```python
# Sử dụng StandardScaler để chuẩn hóa cột 'HCM' và 'LA'
from sklearn .preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['HCM', 'LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM', 'LA']].columns)
print(scaled_df)
```
