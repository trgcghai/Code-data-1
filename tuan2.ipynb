```python
# Nhập thư viện pandas để xử lý dữ liệu và seaborn để trực quan hóa dữ liệu.
import pandas as pd 
import seaborn as sns

# Đọc tệp CSV vào DataFrame, sử dụng 'utf-8' làm mã hóa và dấu phẩy ',' làm dấu phân cách.
df = pd.read_csv('dulieuxettuyendaihoc.csv', header=0, delimiter=',', encoding='utf-8')
# Chọn một số cột nhất định từ DataFrame để sử dụng trong phân tích.
df = df[['GT','DT','KV','KT','M1','M2','M3','NGAYTHI','DINHHUONGNGHENGHIEP']]
df
```

```python
# Hiển thị thông tin tóm tắt về DataFrame bao gồm kiểu dữ liệu và số lượng giá trị không null.
df.info()
```

```python
# Đổi tên các cột 'NGAYTHI' thành 'NT' và 'DINHHUONGNGHIEP' thành 'NGHE'.
df.rename(columns={'NGAYTHI':'NT', 'DINHHUONGNGHENGHIEP':'NGHE'}, inplace=True)
df.columns
```

```python
# Hiển thị 5 dòng đầu tiên của DataFrame.
df.head(5)
```

```python
# Xóa các hàng mà tất cả các giá trị đều là NaN.
df.dropna(how='all', inplace=True)
df
```

```python
# Xóa các hàng trùng lặp trong DataFrame.
df.drop_duplicates(inplace=True)
df
```

```python
# Nhập thư viện matplotlib để tạo biểu đồ.
import matplotlib.pyplot as plt
# Tạo biểu đồ nhiệt (heatmap) để hiển thị các giá trị thiếu trong DataFrame.
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(), cmap='YlGnBu', cbar_kws={'label':'Dữ liệu thiếu'})
```

```python
# Lưu biểu đồ dưới dạng tệp ảnh 'missingdata.png' với độ phân giải 100 dpi.
plt.savefig('missingdata.png', dpi=100)
```

```python
# Hiển thị biểu đồ.
plt.show()
```

```python
# Điền giá trị 'KINH' cho các giá trị bị thiếu trong cột 'DT'.
df['DT'].fillna('KINH', inplace=True)
df
```

```python
# Điền giá trị trung bình cho các giá trị bị thiếu trong cột 'M1'.
df['M1'].fillna(df['M1'].mean(), inplace=True)
df
```

```python
# Điền giá trị trung vị cho các giá trị bị thiếu trong cột 'M2'.
df['M2'].fillna(df['M2'].median(), inplace=True)
df
```

```python
# Điền giá trị 0 cho các giá trị bị thiếu trong cột 'M3'.
df['M3'].fillna(0, inplace=True)
df
```

```python
# Tạo cột mới 'TBM' là trung bình của các cột 'M1', 'M2' và 'M3'.
df['TBM'] = (df['M1'] + df['M2'] + df['M3']) / 3
df
```

```python
# Gán giá trị 'FAIL' cho cột 'XL' nếu 'TBM' nhỏ hơn 5.0.
df.loc[df['TBM'] < 5.0, 'XL'] = 'FAIL'

# Gán giá trị 'FAIR' cho cột 'XL' nếu 'TBM' từ 5.0 đến dưới 7.0.
df.loc[(df['TBM'] >= 5.0) & (df['TBM'] < 7.0), 'XL'] = 'FAIR'

# Gán giá trị 'GOOD' cho cột 'XL' nếu 'TBM' từ 7.0 đến dưới 9.0.
df.loc[(df['TBM'] >= 7.0) & (df['TBM'] < 9.0), 'XL'] = 'GOOD'

# Gán giá trị 'EXCEL' cho cột 'XL' nếu 'TBM' lớn hơn hoặc bằng 9.0.
df.loc[df['TBM'] >= 9.0, 'XL'] = 'EXCEL'

# Hiển thị DataFrame sau khi đã xử lý.
df
```

```python
# Tạo một dictionary để ánh xạ các giá trị của cột 'KT' sang một nhóm mới.
dict_map = {
    'A1': 'G1', 'C': 'G3', 'D1': 'G3', 'A': 'G1', 'B': 'G2'
}

# Tạo cột mới 'NHOM' bằng cách ánh xạ giá trị của cột 'KT' với dictionary.
df['NHOM'] = df['KT'].map(dict_map)

# Hiển thị DataFrame sau khi đã xử lý.
df
```

```python
# Định nghĩa hàm 'fplus' để tính toán giá trị dựa trên điều kiện của cột 'NHOM' và 'TBM'.
def fplus(x, y):
    if (x == 'G1' or x == 'G2') and (y >= 4.0):
        return 1.0
    else:
        return 0.0

# Tạo cột mới 'PLUS' bằng cách áp dụng hàm 'fplus' cho từng hàng của DataFrame.
df['PLUS'] = list(map(fplus, df['NHOM'], df['TBM']))

# Hiển thị DataFrame sau khi đã xử lý.
df
```

```python
# Nhóm dữ liệu theo khu vực ('KV') và tính tổng giá trị của cột 'PLUS' trong mỗi nhóm.
gb = df.groupby(['KV'])['PLUS'].agg(['sum'])

# Tạo nhãn cho biểu đồ hình tròn từ các khu vực ('KV').
labels = gb.index

# Tạo danh sách dữ liệu từ tổng giá trị của cột 'PLUS' trong mỗi nhóm.
data = list(gb['sum'])

# Tạo bảng màu pastel cho biểu đồ.
colors = sns.color_palette('pastel')

# Vẽ biểu đồ hình tròn với dữ liệu đã tính toán.
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.show()
```
