```python
#Transformation Technique: biến dữ liệu từ dạng nguyên bản đọc được sang dạng máy
# Binding
# One-hot Encoding
# Functions,....
# => Function prepareation


import pandas as pd

# Create DataFrame
df = pd.DataFrame({'Math': [4, 4, 7, 8, 12, 13, 15, 18, 22, 23, 23, 25],
                   'Eng': [2, 5, 4, 7, 7, 8, 5, 4, 5, 11, 13, 8]})
df.shape

df['bin1'] = pd.qcut(df['Math'], q=5)
df
```

```python
df.shape
```

```python
# Binding
# chia làm 5 khoảng có số lượng phần tử trong từng khoảng xấp xỉ nhau

df['bin1'] = pd.qcut(df['Math'], q=5)
df
# đếm số lượng mỗi phần tử
df['bin1'].value_counts()
```

```python
# cũng là binding nhưng chia theo phân vị 
df['bin2'] = pd.qcut(df['Math'], q=[0, .2, .4, .6, .8, 1])


df['bin2'].value_counts()
```

```python
# được đánh nhãn theo labels
df['bin3'] = pd.qcut(df['Math'],
                           q=[0, .2, .4, .6, .8, 1],
                           labels=['A', 'B', 'C', 'D', 'E'])
df

df['bin3'].value_counts()
```

```python
# chia theo khoảng độ lớn (0, 5 => >=0 && <= 5)
df['bin4'] = pd.cut(df['Eng'],
                      bins=[0, 5, 8, float('Inf')],
                      labels=['Bad', 'OK', 'Good'])
df
```

```python
# functions
# Normalization 
# Standardization
# popularpuction


import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0] # lương HCM

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0] # lương LA

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

```python
#Normalization: 
# Max Absoulute: biến dữ liệu cũ thành mới nằm trong [-1,1]
# (x(cũ)) / |max|
# |max| : max absoulute


max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la

df
```

```python
# |MAX| của HCM
max_la = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_la

df
```

```python
# chia dữ liệu theo cột

from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
# Hãy cho biết với mức lương HCM và LA lần lượt là 4.5 và 5.75
vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
# Hãy cho biết giá trị lương của HCM và LA trước khi scale
# nếu biết trước giá trị sau khi scale lần lượt là 0

vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

```python
# Hướng dẫn min-max scale về đoạn [0,1]

min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
min_la = df['HCM'].min()
max_la = df['HCM'].max()
df['mmcHCM'] = (df['HCM'] - min_la) / (max_la - min_la)

df
```

```python
# dùng hàm
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
# Hãy cho biết với mức lương HCM và LA lần lượt là 4.5 và 5.75
vSc = scaler.transform([[11.0,4.5]])
vSc
```

```python
#Gỉa sử có giá trị lần lượt là 0.65 và 1.9
vR = scaler.inverse_transform([[0.65,1.9]])
vR
```

```python
#Standardrattion: tính ra độ đo giữa các data độc lập


# Tính Z-Score
mean_la = df['LA'].mean()
std_la = df['LA'].std()
df['zLA'] = (df['LA'] - mean_la) / std_la

mean_la = df['HCM'].mean()
std_la = df['HCM'].std()
df['zHCM'] = (df['HCM'] - mean_la) / std_la

df
```
